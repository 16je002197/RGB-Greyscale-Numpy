```python
from glob import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array(glob("Train/*"))
X_test = np.array(glob("Test/*"))
print(X_train.shape)
print(X_test.shape)
```

```python
def initialize_parameters():
    W = np.random.randn(1,1,3) * 0.01
    b = np.zeros(shape=(1,1,3))
    parameters = {"W": W,
                  "b": b}
    
    return parameters
```

```python
def forward_propagation(X, parameters):
    W = parameters['W']
    b = parameters['b']
    A =  np.sum(X*W+b, axis=2)
    return A
```

```python
def compute_cost(A, Y):
    cost = ((A - Y) ** 2).mean()
    return cost
```

```python
def backward_propagation(parameters, A, X, Y):
  
    W = parameters['W']
    b = parameters['b']
    
    dA = (2/(X.shape[0]*X.shape[1]))*(A-Y) # H x W 
    dA = dA.reshape([1,-1])
    X = X.reshape([-1,3])
    dW = np.dot(dA,X).reshape([1,1,3]) #1 x 1 x 3
    db = np.dot(dA, np.ones((dA.shape[1],3)))
    
    grads = {"dW": dW,
             "db": db}
    
    return grads
```

```python
def update_parameters(parameters, grads, learning_rate=0.05):

    W = parameters['W']
    b = parameters['b']
    
    dW = grads['dW']
    db = grads['db']
    
    W = W - learning_rate * dW
    b = b - learning_rate * db
    
    parameters = {"W": W,
                  "b": b}
    
    return parameters
```

```python
def predict(parameters, X):
    
    A = forward_propagation(X, parameters)
    return A
```

```python
def nn_model(X, num_epochs=2):
    
    costs = []
    parameters = initialize_parameters()
    W = parameters['W']
    b = parameters['b']
    
    for i in range(num_epochs):
        for img_path in X:
            x = cv2.imread(img_path) 
            x = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
            y = cv2.cvtColor(x, cv2.COLOR_RGB2GRAY)

            x = x/255
            y = y/255
            
            A = forward_propagation(x, parameters)
            cost = compute_cost(A, y)
            
            grads = backward_propagation(parameters, A, x, y)
            parameters = update_parameters(parameters, grads)
            costs.append(cost)
        print ("Cost after epoch %i/%i: %f" % (i+1,num_epochs, cost))
        plt.plot(costs)
        plt.xlabel('Cost after ith example')
        plt.ylabel('Cost')
        plt.show()
    return parameters
```

```python

parameters = nn_model(X_train)
```

```python
def convert(img, parameters):
    greyimg = forward_propagation(img,parameters)
    return greyimg
```

```python
#Testing part
for img_path in X_test:
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    greyimg = convert(img,parameters)
    print('Colored: ')
    plt.imshow(img)
    plt.show()
    
    print('Ground Truth Greyscale')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY),cmap='gray')
    plt.show()
    
    print('Predicted Greyscale')
    plt.imshow(greyimg,cmap='gray')
    plt.show()
```

```python
#Difference between random weights and trained weights by comparing with ground truth
rand_para = initialize_parameters()
idx = np.random.randint(X_test.shape[0])

test_img_path = X_test[idx]
test_img = cv2.imread(test_img_path)
test_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB)
greyimg = convert(test_img,parameters)
randgreyimg = convert(test_img,rand_para)
print('Colored: ')
plt.imshow(test_img)
plt.show()
    
print('Ground Truth Greyscale')
plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_RGB2GRAY),cmap='gray')
plt.show()
    
print('Predicted Greyscale')
plt.imshow(greyimg,cmap='gray')
plt.show()

print('Random weighted image')
plt.imshow(randgreyimg,cmap='gray')
plt.show()
```
